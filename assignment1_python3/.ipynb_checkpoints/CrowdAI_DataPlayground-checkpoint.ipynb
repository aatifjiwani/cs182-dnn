```python
class CrowdAIDataset(Dataset):
    def __init__(self, baseDir, annotation):
        self.annoDir = "{}/annotations/{}.json".format(baseDir, annotation)
        self.imgDir = "{}/images/".format(baseDir)
        
        self.coco = COCO(self.annoDir)
        self.cats = coco.loadCats(coco.getCatIds())
        self.catNames = [cat['name'] for cat in cats]
        self.primaryCategories = None
        
        print(f"COCO categories: {self.catNames}")
        
    def setCategories(self, cat):
        self.primaryCategories = cat
        self.primaryCatIds = self.coco.getCatIds(catNms=cat)
        self.imgIds = self.coco.getImgIds(catIds = self.primaryCatIds)
        
        print(f"Set primary categories to {cat}")
        
    def __len__(self):
        return len(self.imgIds)
    
    def __getitem__(self, index):
        batch = self.coco.loadImgs(self.imgIds[index])
        sample =  {"images": getImage(batch, True), "masks": getMask(batch)}
        
    def getImage(self, batch, multichannel):
        width = batch[0]["width"]
        height = batch[0]["height"]
        numBatch = len(batch)
        channels = 3 if multichannel else 1
        
        tensor = np.empty([numBatch, width, height, channels])
        
        for i in range(0, numBatch):
            img = batch[i]['file_name']
            tensor[i, :, :, :] = io.imread('{}{}'.format(self.imgDir,img))
            
        return tensor if len(tensor) > 1 else tensor[0]
    
    def getMask(self, batch):
        numBatch, width, height, channels = len(batch), batch[0]["width"], batch[0]["height"], \
            len(self.primaryCategories) + 1  ##Background category
        
        maskTensor = np.empty([numBatch, width, height, channels])
        for i in range(0, numBatch):
            mask = np.empty([channels, width, height])
            imgId = batch[i]['id']
            annIds = coco.getAnnIds(imgIds=imgId, catIds=self.primaryCatIds, iscrowd=None)
            anns = coco.loadAnns(annIds)
            for j in range(0, len(anns)):
                ann = anns[j]
                toChannel = self.primaryCatIds.index(ann["category_id"]) + 1
                mask[toChannel, :, :] = np.maximum(mask[toChannel, :, :], self.coco.annToMask(ann))
                
            mask[0, :, :] = np.full((width, height), 1)
            for k in range(1, len(mask)):
                mask[0,:,:] = mask[0,:,:] - mask[k, :, :]
                
            maskTensor[i, :, :, :] = mask.T
            
        return maskTensor if len(maskTensor) > 1 else maskTensor[0]
            
            
            
        
```

```python
class COCOAnnotationToMask:
    """
        Directory should be in the structure:
        /root
            COCOAnnotationToMask.py
            baseDir/
                /annotations
                    annotations.json
                /images
                    ... list of images from dataset
                /masks (optional)
            
    """
    def __init__(self, baseDir, annotationFile):
        self.rootDir = baseDir
        self.annoDir = "{}/annotations/{}.json".format(baseDir, annotationFile)
        self.imgDir = "{}/images/".format(baseDir)
        
        self.coco = COCO(self.annoDir)
        self.cats = coco.loadCats(coco.getCatIds())
        self.catNames = [cat['name'] for cat in cats]
        self.primaryCategories = None
        
        if 'masks' not in os.listdir(baseDir):
            os.mkdir(f"{baseDir}/masks")
            
        
        print(f"COCO categories: {self.catNames}")
    
    def setCategories(self, cat):
        self.primaryCategories = cat
        self.primaryCatIds = self.coco.getCatIds(catNms=cat)
        self.imgIds = self.coco.getImgIds(catIds = self.primaryCatIds)
        
        print(f"Set primary categories to {cat}")
        
    def convertAllToMask(self):
        self.convertToMask(0, len(self.imgIds))
        
    def convertToMask(self, start, end):
        imgMeta = self.coco.loadImgs(self.imgIds[start:end])
        for img in imgMeta:
            width, height, channels = img["width"], img["height"], len(self.primaryCategories) + 1  ##Background category
            mask = np.empty([channels, width, height])
            imgId = img['id']
            annIds = self.coco.getAnnIds(imgIds=imgId, catIds=self.primaryCatIds, iscrowd=None)
            anns = self.coco.loadAnns(annIds)
            for j in range(0, len(anns)):
                ann = anns[j]
                toChannel = self.primaryCatIds.index(ann["category_id"]) + 1
                mask[toChannel, :, :] = np.maximum(mask[toChannel, :, :], self.coco.annToMask(ann))
                
            mask[0, :, :] = np.full((width, height), 1)
            for k in range(1, len(mask)):
                mask[0,:,:] = mask[0,:,:] - mask[k, :, :]
            
            fileName = img['file_name'][0:img['file_name'].index('.')]
            np.save(f"{self.rootDir}/masks/{fileName}", mask)
            break
            
            
            
            
            
```
